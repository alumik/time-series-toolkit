```python
%load_ext autoreload
%autoreload 2


import tskit
import seaborn as sns

sns.set_style('darkgrid')
```

```python
ts_mod = tskit.generator.random_walk(length=100)
tskit.plot(ts_mod)
```

```python
ts_base = tskit.generator.sine(length=100)
tskit.plot(ts_base)
```

```python
ts = tskit.transform.combine([ts_base, ts_mod], weights=[1.0, 0.5], standardize_idx=[1])
tskit.plot(ts)
```

```python
ts.smooth('ewma', alpha=0.5)
tskit.plot(ts)
```

```python
ts.to_shapelet(alpha=5)
tskit.plot(ts)
```

```python
ts = ts.tile(10)
tskit.plot(ts)
```

```python
ts.add_noise('gaussian')
tskit.plot(ts)
```
