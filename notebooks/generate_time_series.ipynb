```python
%load_ext autoreload
%autoreload 2


import tskit
import seaborn as sns

sns.set_style('darkgrid')
```

```python
ts = (
    tskit.TimeSeries.from_generators(['sine', 'random_walk'], weights=[1.0, 0.5], standardize_idx=[1], length=100)
    .smooth('ewma', alpha=0.5)
    .to_shapelet(alpha=5)
    .tile(10)
    .add_noise('gaussian', amplitude=0.3, clip=[-0.2, 0.2])
)
tskit.plot(ts)
```
