```python
%load_ext autoreload
%autoreload 2


import tskit
import seaborn as sns

sns.set_style('darkgrid')
```

```python
ts = (
    tskit.TimeSeries.from_generators(
        generators=['sine', 'random_walk'],
        generator_args=[{'amplitude': 0.5}, {}],
        weights=[1.0, 0.5],
        standardize_idx=[1],
        length=100,
    )
    .smooth('ewma', alpha=0.5)
    .to_shapelet(alpha=5)
    .tile(10)
    .add_noise('gaussian', amplitude=0.3, clip=[-0.2, 0.2])
)
tskit.plot(ts)
```

```python
ts = (
    tskit.TimeSeries.from_config({
        'generators': [
            {
                'name': 'sine',
                'weight': 1.0,
                'args': {
                    'amplitude': 0.5,
                },
            },
            {
                'name': 'random_walk',
                'weight': 0.5,
                'standardize': True,
            },
        ],
        'length': 100,
    })
    .smooth('ewma', alpha=0.5)
    .to_shapelet(alpha=5)
    .tile(10)
    .add_noise('gaussian', amplitude=0.3, clip=[-0.2, 0.2])
)
tskit.plot(ts)
```
