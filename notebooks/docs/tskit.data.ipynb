# tskit.data

## tskit.data.TimeSeries

### tskit.data.TimeSeries.__init__()

Create a new TimeSeries object from integer index values.

```python
tskit.TimeSeries(
    index=[0, 2, 7],
    values=[0, 1, 2],
).plot()
```

Create a new TimeSeries object from `pd.DateTimeIndex` values.

```python
import datetime

tskit.TimeSeries(
    index=pd.DatetimeIndex([
        datetime.datetime(2000, 1, 3),
        datetime.datetime(2000, 1, 1),
        datetime.datetime(2000, 1, 4),
    ]),
    values=[0, 1, 2],
    name='foo',
).plot()
```

### tskit.data.TimeSeries.from_generators()

```python
tskit.TimeSeries.from_generators(
    generators=['sine', 'random_walk'],
    generator_args=[{'amplitude': 0.5}, {}],
    weights=[1.0, 0.5],
    standardize_idx=[1],
    length=100,
).plot()
```

### tskit.data.TimeSeries.from_config()

```python
tskit.TimeSeries.from_config({
    'generators': [
        {
            'name': 'sine',
            'weight': 1.0,
            'args': {
                'amplitude': 0.5,
            },
        },
        {
            'name': 'random_walk',
            'weight': 0.5,
            'standardize': True,
        },
    ],
    'length': 100,
}).plot()
```
