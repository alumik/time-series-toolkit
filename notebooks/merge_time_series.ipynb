## Merge two sin wave

```python
ts_sin_1 = tskit.generator.sine(length=100, period=100, name='sin_1')
tskit.plot(ts_sin_1)
ts_sin_2 = tskit.generator.sine(length=100, phase=20, amplitude=2, period=50, name='sin_2')
tskit.plot(ts_sin_2)
ts_sin_3 = tskit.generator.sine(length=100, phase=20, period=30, name='sin_3')
tskit.plot(ts_sin_3)
```

```python
ts_mix_1 = tskit.transform.add([ts_sin_1, ts_sin_2, ts_sin_3], weights=[1.0, 0.5, 0.8])
tskit.plot(ts_mix_1)
```
