```python
%load_ext autoreload
%autoreload 2


import tskit
import seaborn as sns

sns.set_style('darkgrid')
```

```python
ts_mod = tskit.generator.random_walk(length=100)
tskit.plot(ts_mod)
```

```python
ts_base = tskit.generator.sine(length=100)
tskit.plot(ts_base)
```

```python
ts = tskit.transform.combine([ts_base, ts_mod], weights=[1.0, 0.5], standardize_idx=[1])
tskit.plot(ts)
```

```python
ts = tskit.smoothing.ewma(ts, alpha=0.5)
tskit.plot(ts)
```

```python
ts = tskit.transform.to_shapelet(ts, alpha=5)
tskit.plot(ts)
```

```python
ts = tskit.transform.tile(ts, 10)
tskit.plot(ts)
```

```python
ts = tskit.noise.add_gaussian_noise(ts)
tskit.plot(ts)
```
